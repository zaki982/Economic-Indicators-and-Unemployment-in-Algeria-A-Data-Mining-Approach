## **Data Preprocessing**

To ensure the dataset is clean and reliable for analysis, we applied a series of preprocessing steps. The goal of this step is to address missing values, detect and handle noise and outliers, and normalize the data for consistency across variables.

## Handling Missing Values

Missing data can lead to inaccurate analysis and biased results. In this dataset, the unemployment value for 1990 was missing. To address this:

1. **Interpolation**: Missing values were estimated based on the trend of nearby data points using linear interpolation.
2. **Filling Remaining Missing Values**: Any remaining `NaN` values after interpolation were filled using the column's mean to ensure no gaps remain in the data.

Below is the implementation of this step:


```python
# Ensure columns 'GDP', 'Inflation', and 'Unemployment' in combined_df are numeric
for column in ['GDP', 'Inflation', 'Unemployment']:
    combined_df[column] = pd.to_numeric(combined_df[column], errors='coerce')  # Convert to numeric, forcing errors to NaN

# Interpolate missing values using linear interpolation
combined_df = combined_df.sort_values('Year')  # Sort by 'Year' to ensure proper interpolation
for column in ['GDP', 'Inflation', 'Unemployment']:
    combined_df[column] = combined_df[column].interpolate(method='linear')  # Fill missing values by interpolation
    combined_df[column] = combined_df[column].fillna(combined_df[column].mean())  # Fill remaining NaNs with the column mean

combined_df.to_csv("combined_data.csv", index=False)

# Display the updated dataset with missing values handled
print(combined_df)

```

## **Data Transformation**

Data transformation is a crucial step in preparing datasets for analysis. It involves converting raw data into a more suitable format for extracting meaningful insights. In this project, we applied data discretization to categorize continuous variables into discrete bins, making them more interpretable and suitable for specific data mining techniques.

### Data Discretization

Discretization is the process of converting continuous numerical data into discrete categories. For this dataset, we applied **equal-width binning**, an unsupervised method that divides the range of each continuous variable into equal-sized bins. This approach ensures that the intervals are uniform in size, providing a simple and consistent categorization.

We performed discretization on the following columns:
- **Inflation**: Binned into three categories: `Low`, `Moderate`, and `High`.
- **GDP**: Binned into three categories: `Low`, `Moderate`, and `High`.
- **Unemployment**: Binned into three categories: `Low`, `Moderate`, and `High`.

The resulting categories allow us to simplify the analysis and make the data more suitable for **association rule mining**. 

Below is the code used for discretization:

```python
# Equal-width binning for categorization
combined_df["Inflation_Category"] = pd.cut(combined_df["Inflation"], bins=3, labels=["Low", "Moderate", "High"])
combined_df["GDP_Category"] = pd.cut(combined_df["GDP"], bins=3, labels=["Low", "Moderate", "High"])
combined_df["Unemployment_Category"] = pd.cut(combined_df["Unemployment"], bins=3, labels=["Low", "Moderate", "High"])

# Save combined CSV file
combined_df.to_csv("combined_data.csv", index=False)
pd.set_option('display.max_columns', None)
# Drop continuous numerical columns and keep categorical ones
final_df = combined_df[["Year", "Inflation_Category", "GDP_Category", "Unemployment_Category"]]
# Save the final DataFrame to a CSV file
final_df.to_csv("final_categorized_data.csv", index=False)
# Display the final DataFrame
print(final_df)
```

# Association Rule Mining on Economic Indicators and Unemployment
Association rule mining uncovers hidden patterns and relationships between economic indicators, offering valuable insights into how these variables interact under various economic conditions.

In this project, we aim to apply association rule mining to uncover connections between **economic indicators** such as inflation, GDP, and unemployment categories. These insights will help us better understand how economic conditions interact with unemployment trends.

## Goal
By identifying strong association rules, we aim to uncover patterns that describe how changes in inflation and GDP relate to unemployment trends.

## Choice of Algorithm: 

### Recommendation: FP-Growth Algorithm
Given the relatively small size of our dataset, **both Apriori and FP-Growth can be used effectively**. However, for efficiency and scalability (in case of future extensions with larger datasets), we recommend using the **FP-Growth algorithm**.

### FP-Growth Algorithm
The FP-Growth algorithm is an alternative to Apriori that constructs a compact **Frequent Pattern Tree (FP-Tree)** and avoids multiple dataset scans. It performs association rule mining more efficiently for large datasets.

## Why FP-Growth for This Dataset?
- The dataset is relatively small, so FP-Growth will process it efficiently without the computational overhead of Apriori.
- FP-Growth avoids the need for generating a large number of candidate itemsets, making it faster and more scalable.
- It provides a robust foundation in case of future expansions or additions to the dataset.


# Step 1: Data Transformation for FP-Growth

The FP-Growth algorithm requires data in a transactional format, where each transaction represents a set of items. 
In this step, we convert each row of the dataset into a list of categorical items.


```python
# Transform each row into a list of transactions
transactions = final_df.apply(lambda row: [
    f"{row['Inflation_Category']}_Inflation",
    f"{row['GDP_Category']}_GDP",
    f"{row['Unemployment_Category']}_Unemployment"
], axis=1).tolist()

# Print the first few transactions
print(transactions[:5])
```

# Step 2: Choosing Minimum Support and Confidence Thresholds

The 15% support threshold was selected to capture patterns that occur frequently enough to be meaningful without being overly restrictive. The 60% confidence threshold ensures rules are reliable predictors while minimizing false positives.

- **Minimum Support**: Specifies how frequently an itemset appears in the dataset. We set it to 15% (0.15) to capture patterns that are not too rare.
- **Minimum Confidence**: Specifies how reliable a rule is in predicting the consequent. We set it to 60% (0.6) for reliability.


```python
# Define thresholds
min_support = 0.15  # 15% of the dataset
min_confidence = 0.6  # 60% confidence

print(f"Minimum Support: {min_support}")
print(f"Minimum Confidence: {min_confidence}")
```

# Step 3: Applying the FP-Growth Algorithm

### Explanation:
The FP-Growth algorithm identifies frequent itemsets without generating all possible combinations (like Apriori does). 
In this step, we use the **mlxtend** library to apply FP-Growth and extract frequent itemsets based on the support threshold.

1. **Why One-Hot Encoding?**  
   FP-Growth requires data in a format where each item is represented as a column with binary values (`1` for presence, `0` for absence). This transformation helps the algorithm efficiently process the data.
2. **What Are Frequent Itemsets?**  
   A frequent itemset is a group of items that occur together in a significant portion of the transactions (defined by `min_support`). For example, if `High_Inflation` and `Low_GDP` appear together in 20% of the transactions, they form a frequent itemset .

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder

# Encode transactions into a one-hot encoded dataframe
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply FP-Growth to find frequent itemsets
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets = fpgrowth(df, min_support=min_support, use_colnames=True)

# Display frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)
```

# Step 4: Generating Association Rules Metric Selection  

For evaluating the rules in our dataset, we chose the following metrics: 

1. **Confidence**:  
   Confidence tells us the conditional probability of the consequent occurring given the antecedent. Higher confidence means the rule is more reliable.  
- **Confidence**: \( \text{Confidence} = \frac{\text{Support(A âˆª B)}}{\text{Support(A)}} \)  

2. **Lift**:  
   Lift measures how much the occurrence of the antecedent increases the likelihood of the consequent compared to random chance. A lift greater than 1 indicates a strong positive association.  
- **Lift**: \( \text{Lift} = \frac{\text{Confidence}}{\text{Support(B)}} \)  

3. **Leverage**:  
   Leverage calculates how much the co-occurrence of antecedent and consequent exceeds what is expected under independence. It quantifies the added value of the rule beyond random chance.  
- **Leverage**: \( \text{Leverage} = \text{Support(A âˆª B)} - \text{Support(A)} \times \text{Support(B)} \)


We selected these metrics as they collectively provide:
- **Confidence** for rule reliability.  
- **Lift** for rule strength compared to randomness.  
- **Leverage** for statistical significance and added value.

The results from these metrics will allow us to identify and prioritize the most meaningful and impactful association rules.

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules with a lower threshold for confidence
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

# Filter for strong relationships with lift > 1 and leverage > 0
rules = rules[(rules['lift'] > 1) & (rules['leverage'] > 0)]

# Sort rules by confidence, lift, and leverage
rules = rules.sort_values(by=['confidence', 'lift', 'leverage'], ascending=[False, False, False])

# Export the generated rules to a CSV file for further analysis
rules.to_csv('association_rules.csv', index=False)

# Display only selected columns with meaningful metrics
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift', 'leverage']])

# For a quick overview, limit the output to the top 10 strongest rules
print("\nStrongest Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift', 'leverage']].head(10))
```
